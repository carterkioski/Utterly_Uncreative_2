```python
# from tensorflow import lite
# import tensorflow as tf
# from tensorflow import keras
# from tensorflow.keras import datasets, layers, models
import numpy as np
import pandas as pd
# import random, os
import shutil
import matplotlib.pyplot as plt
from matplotlib.image import imread
# from tensorflow.keras.preprocessing.image import ImageDataGenerator
# from tensorflow.keras.metrics import categorical_accuracy, AUC
# from sklearn.model_selection import train_test_split

reload_history=pd.read_csv("100_history_data.csv", index_col=0)
reload_history

new_history = reload_history.to_dict("split")

F_history = dict(zip(new_history["index"], new_history["data"]))
```

```python
# plt.subplot(1,3,1)
plt.plot(F_history['acc'])
plt.plot(F_history['val_acc'])
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validate'], loc='center right')
plt.savefig('plot_1.png', dpi=300, bbox_inches='tight')
```

```python
# plt.subplot(1,3,3)
plt.plot(F_history['loss'])
plt.plot(F_history['val_loss'])
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validate'], loc='center right')
plt.savefig('plot_2.png', dpi=300, bbox_inches='tight')
```
